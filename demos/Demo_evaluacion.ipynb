## Modelo 6 clases

```python
import torch
```

```python
def load_checkpoint(model, filepath):
    checkpoint_state_dict = torch.load(filepath)
    model.load_state_dict(checkpoint_state_dict)
    
    return model
```

```python
#ruta checkpoint
ruta_modelo_6_clases = '/content/drive/MyDrive/MDS/2022-2/Procesamiento avanzado de imaÃÅgenes/Proyecto/ViT_6.pth'

#instanciar modelo
vit_6_clases = ViT(n_classes = 6,
                   emb_size = 768,
                   num_heads = 12)

#cargar modelo pre-entrenado
modelo_6_clases = load_checkpoint(vit_6_clases, ruta_modelo_6_clases).to(device)
```

### Generar predicci√≥n y evaluaci√≥n


```python
device = ('cuda' if torch.cuda.is_available() else 'cpu')
labels_6, predictions_6 = predicciones_modelo(modelo_6_clases,
                                              device,
                                              test_loader)
```
